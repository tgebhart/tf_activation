```python
import os
import sys
import progressbar

import dionysus as dion
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from tf_activation.functions.persistence_landscape import persistence_landscape, average_landscape
import seaborn as sns
%load_ext autoreload
%autoreload 2
%matplotlib notebook
```

```python
base_loc = '/home/gebha095/projects/tf_activation/logdir/data/experiments/filtrations/cff_wc15-5-32_wfc11024_seed1_normalstd2p0_relu/p99_mnist_SGD1p0e-04_100epochs_128batch'
t_t_type = 'train'
```

```python
classes = list(range(10))
```

```python
diagram_df = []

for c in classes:
    print('Class: {}'.format(c))
    c_loc = os.path.join(base_loc,str(c))
    for filtration_file in os.listdir(c_loc):
        
        if filtration_file.find(t_t_type) != -1:
            r = np.genfromtxt(os.path.join(c_loc, filtration_file), delimiter=',')

            max_eps = r[0]
            f = dion.Filtration()
            for i in range(r.shape[0]):
                f.append(dion.Simplex([i+1], r[i]))
                f.append(dion.Simplex([2*(i+1)], r[i]))
                f.append(dion.Simplex([i+1, 2*(i+1)], r[i]))
            f.sort(reverse=True)

            m = dion.homology_persistence(f)
            dgms = dion.init_diagrams(m, f)

            ex = filtration_file[:filtration_file.find('_')]
            diagram_df.append({'class': c, 'diagrams':dgms, 'example':ex, 'max_filtration':max_eps})
```

```python
diagram_df = pd.DataFrame(diagram_df)
```

```python
H = 0
EVERY = 1
```

```python
def remove_infs(dgm, m=0):
    return dion.Diagram([(p.birth, p.death) if p.death < float('Inf') else (p.birth, m) for p in dgm])
```

```python
dist_dim = diagram_df.shape[0]
print('Number rows: {}'.format(dist_dim))
dists = np.zeros(shape=(dist_dim//EVERY+1,dist_dim//EVERY+1))
cols = np.empty(shape=(dist_dim//EVERY))

bar = progressbar.ProgressBar(maxval=dist_dim, \
    widgets=[progressbar.Bar('=', '[', ']'), ' ', progressbar.Percentage()])
bar.start()
for i in range(0,dist_dim,EVERY):
    print(i, '/', dist_dim)
    for j in range(0,dist_dim,EVERY):
        if i != j:
            d1 = remove_infs(diagram_df['diagrams'].iloc[i][H],m=0)
            d2 = remove_infs(diagram_df['diagrams'].iloc[j][H],m=0)
            dists[i//EVERY,j//EVERY] = dion.wasserstein_distance(d1, d2, q=2)
    cols[i//EVERY] = diagram_df['class'].iloc[i]
```

```python
def plot_landscape(L, cls, every=100, max=25):

    ax = plt.subplot(111,projection='3d')
    # ax = plt.subplot()
    for i in range(0, max):

        xs = []
        ys = []
        zs = []
        for l in L[i]:
            if abs(l[0]) != float('Inf') and abs(l[1]) != float('Inf'):
                xs.append(l[0])
                ys.append(i)
                zs.append(l[1])

        plt.plot(xs, ys, zs)

    # now plot both limits against eachother

    # ax.set_aspect('equal')
    plt.xlabel('Birth Time')
    plt.ylabel('Death Time')
    plt.title('Class {}'.format(cls))
    plt.show()
```

```python
def plot_diagram(d, cls):

    ax = plt.subplot()
    diag = d[~np.isinf(d[:,1])]
    ax.scatter(diag[:,0], diag[:,1], s=25, c=(diag[:,0] - diag[:,1])**2, cmap=plt.cm.coolwarm, zorder=10)
    lims = [
        np.min(0.85*diag[:,0]),  # min of both axes
        np.max(1.15*diag[:,1]),  # max of both axes
    ]

    # now plot both limits against eachother

    ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
    ax.set_aspect('equal')
    ax.set_xlim(lims)
    ax.set_ylim(lims)
    plt.xlabel('Birth Time')
    plt.ylabel('Death Time')
    plt.title('Class {}'.format(cls))

    plt.show()
    
def diag_to_numpy(diag):
    ret = np.empty(shape=(len(diag),2))
    i = 0
    for p in diag:
        ret[i,0] = p.birth
        ret[i,1] = p.death
        i += 1
    return ret
```
