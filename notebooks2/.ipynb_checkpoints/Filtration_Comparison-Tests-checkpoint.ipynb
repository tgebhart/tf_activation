```python
import os
import sys
import progressbar

import dionysus as dion
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_loc = '/home/gebha095/projects/tf_activation/logdir/data/experiments/filtrations/cff_wc15-5-32_wfc11024_seed1_uniformstd2p0_relu/p99_mnist_SGD1p0e-04_100epochs_128batch'
t_t_type = 'test'
```

```python
classes = list(range(10))
```

```python
diagram_df = []

for c in classes:
    print('Class: {}'.format(c))
    c_loc = os.path.join(base_loc,str(c))
    for filtration_file in os.listdir(c_loc):
        
        if filtration_file.find('test') != -1:
            r = np.genfromtxt(os.path.join(c_loc, filtration_file), delimiter=',')

            max_eps = r[0]
            f = dion.Filtration()
            for i in range(r.shape[0]):
                f.append(dion.Simplex([i+1], r[i]))
                f.append(dion.Simplex([2*(i+1)], r[i]))
                f.append(dion.Simplex([i+1, 2*(i+1)], r[i]))
            f.sort(reverse=True)

            m = dion.homology_persistence(f)
            dgms = dion.init_diagrams(m, f)

            ex = filtration_file[:filtration_file.find('_')]
            diagram_df.append({'class': c, 'diagrams':dgms, 'example':ex, 'max_filtration':max_eps})
```

```python
diagram_df = pd.DataFrame(diagram_df)
```

```python
diagram_df.head()
```

```python
H = 0
EVERY = 20
```

```python
def remove_infs(dgm, m=0):
    return dion.Diagram([(p.birth, p.death) if p.death < float('Inf') else (p.birth, m) for p in dgm])
```

```python
dist_dim = diagram_df.shape[0]
print('Number rows: {}'.format(dist_dim))
dists = np.zeros(shape=(dist_dim//EVERY+1,dist_dim//EVERY+1))
cols = np.empty(shape=(dist_dim//EVERY))

bar = progressbar.ProgressBar(maxval=dist_dim, \
    widgets=[progressbar.Bar('=', '[', ']'), ' ', progressbar.Percentage()])
bar.start()
for i in range(0,dist_dim,EVERY):
    print(i, '/', dist_dim)
    for j in range(0,dist_dim,EVERY):
        if i != j:
            print(j)
            d1 = remove_infs(diagram_df['diagrams'].iloc[i][H], m=0)
            d2 = remove_infs(diagram_df['diagrams'].iloc[j][H], m=0)
            dists[i//EVERY,j//EVERY] = dion.wasserstein_distance(d1, d2, q=2)
    cols[i//EVERY] = diagram_df['class'].iloc[i]
```

```python
df = pd.DataFrame(dists)
# dists_df = df.drop(27).drop(27, axis=1)
sns.heatmap(df)
```

```python
from tf_activation.models.cff import CFF_Model
model = CFF_Model(W_conv1=[5,5,1,32])

model_directory = '/home/gebha095/projects/tf_activation/logdir/models/cff_wc15-5-32_wfc11024_seed1_normalstd2p0_relu'
model.infer_graph_parameters_from_filename(model_directory)
trained_model = 'mnist_SGD1p0e-04_100epochs_128batch/model.ckpt'
training_parameters = model.infer_training_parameters_from_filename(trained_model)
```

```python
tf.reset_default_graph()

    # Create the model
x = tf.placeholder(tf.float32, [None, 784])

# Define loss and optimizer
y_ = tf.placeholder(tf.float32, [None, 10])

# Build the graph for the deep net
net = model.build_graph(x)
if not model.implements_dropout:
    keep_prob = tf.placeholder(tf.float32)

logits = model.get_output_layer()
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_, logits=logits))
train_step = model.get_train_step(cross_entropy, training_parameters['optimizer'], lr=training_parameters['learning_rate'])
correct_prediction = tf.equal(tf.argmax(model.get_output_layer(), 1), tf.argmax(y_, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

saver = tf.train.Saver()
    
    
    
```

```python
X = np.take(mnist.test.images, big_nums, axis=0)
Y = np.take(mnist.test.labels, big_nums, axis=0)
with tf.Session() as sess:
    
    saver.restore(sess, os.path.join(model_directory, trained_model))
    
    acc = accuracy.eval(feed_dict={
                x: X, y_: Y, keep_prob: 1.0})

    loss = cross_entropy.eval(feed_dict={x: X, y_: Y, keep_prob:1.0})
    
    print('Accuracy: {} , Loss: {}'.format(acc,loss))
```
